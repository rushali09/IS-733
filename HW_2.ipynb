```python
# 1)B)
#importing necessary libraries
#Code to build Naive Bayes
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report
```

```python
# loading dataset
df= pd.read_csv("hamspam.csv")
df.head()
```

```python
# Convert categorical features to numerical values
df["Contains Link"] = df["Contains Link"].map({"Yes": 1, "No": 0})
df["Contains Money Words"] = df["Contains Money Words"].map({"Yes": 1, "No": 0})
df["Length"] = df["Length"].map({"Long": 1, "Short": 0})
df["Class"] = df["Class"].map({"Spam": 1, "Ham": 0})
```

```python
# Define features and target
x = df[["Contains Link", "Contains Money Words", "Length"]]
y = df["Class"]
```

```python
# Split data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Train a Naïve Bayes classifier
nb_model = MultinomialNB()
nb_model.fit(x_train, y_train)
```

```python
# Predictions
y_pred = nb_model.predict(x_test)
y_pred
```

```python
# Evaluate model performance
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
accuracy
report

# Accuracy is 52%
```

```python
#Code to build KNN
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
df= pd.read_csv("hamspam.csv")
df.head()
```

```python
# Convert categorical features to numerical values
df["Contains Link"] = df["Contains Link"].map({"Yes": 1, "No": 0})
df["Contains Money Words"] = df["Contains Money Words"].map({"Yes": 1, "No": 0})
df["Length"] = df["Length"].map({"Long": 1, "Short": 0})
df["Class"] = df["Class"].map({"Spam": 1, "Ham": 0})

# Define features and target
x = df[["Contains Link", "Contains Money Words", "Length"]]
y = df["Class"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Initialize KNN classifier with k=2 (default choice)

knn_model = KNeighborsClassifier(n_neighbors=2)

knn_model.fit(x_train, y_train)
```

```python
# Predictions
y_pred = knn_model.predict(x_test)
y_pred
```

```python
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
accuracy
report

#accuracy is 53%
```

```python
# Task2: Understanding ROC and AUC 
# 2a: Create a ROC (with AI assistant/Excel ) (Refer to roc_data.csv)
# Step1: Given the threshold (0.95,0.90,0.85,0.80,0.75,0.70), derive True Positive and False Positive 
# Step2: Calculate the True Positive Rate (TPR) and False Positive Rate (FPR), enter the values into the sheet 
# Step3: plot the set points (FRP, TPR) on the ROC diagram 



import pandas as pd
DataFrame= pd.read_csv("roc_data.csv")

# As given in the question
thresholds = [0.95, 0.90, 0.85, 0.80, 0.75, 0.70]

# Calculating total positives and total negatives

total_positives = sum(DataFrame["True_Label"] == 1)
total_negatives = sum(DataFrame["True_Label"] == 0)
results = []

# Here i refers to threshold
for i in thresholds:
    DataFrame["Predicted_Label"] = (DataFrame["Prediction"] >= i).astype(int)
    
    TP = sum((DataFrame["Predicted_Label"] == 1) & (DataFrame["True_Label"] == 1))
    FP = sum((DataFrame["Predicted_Label"] == 1) & (DataFrame["True_Label"] == 0))
    TN = sum((DataFrame["Predicted_Label"] == 0) & (DataFrame["True_Label"] == 0))
    FN = sum((DataFrame["Predicted_Label"] == 0) & (DataFrame["True_Label"] == 1))

    TPR = TP / total_positives if total_positives > 0 else 0
    FPR = FP / total_negatives if total_negatives > 0 else 0
    #calculating total count
    total_count = TP + FP + TN + FN
    print(f"Threshold: {i} -> TP: {TP}, FP: {FP}, TN: {TN}, FN: {FN}, Total: {total_count}")

    results.append([i, TP, FP, TN, FN, TPR, FPR])

results
new_dataset = pd.DataFrame(results, columns=["i", "TP", "FP", "TN", "FN", "TPR", "FPR"])
new_dataset
```

```python
# solution for part 3
#Question 2a - Step 3 Solution Code
import pandas as pd
import matplotlib.pyplot as plt


dataframe = pd.read_csv("roc_data.csv")
# As given in the question
thresholds = [0.95, 0.90, 0.85, 0.80, 0.75, 0.70]

tpr_values = []
fpr_values = []

# Calculating total positives and total negatives
total_positives = sum(dataframe["True_Label"] == 1)
total_negatives = sum(dataframe["True_Label"] == 0)

for i in thresholds:
    dataframe["Predicted_Label"] = (dataframe["Prediction"] >= i).astype(int)
    
    TP = sum((dataframe["Predicted_Label"] == 1) & (dataframe["True_Label"] == 1))
    FP = sum((dataframe["Predicted_Label"] == 1) & (dataframe["True_Label"] == 0))
    
    TPR = TP / total_positives if total_positives > 0 else 0
    FPR = FP / total_negatives if total_negatives > 0 else 0
    
    tpr_values.append(TPR)
    fpr_values.append(FPR)


plt.figure(figsize=(5, 6))
plt.plot(fpr_values, tpr_values, marker='o', color='g', label=' Then ROC Curve')
plt.plot([0, 1], [0, 1], linestyle='-', color='gray', label='Random Classifier')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("The ROC Curve")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 2b. Write code (with AI assistant) to fit the model using your favorite classifier 
# (NB, KNN, or Decision tree); using the hamspam.csv, ask to output an ROC curve and AUC score. 
# (Hint: if you fit a decision tree, you might want to reduce max_depth) 


# importing necessary libaries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import roc_curve, auc

hamspam_df= pd.read_csv("hamspam.csv")
# Encode categorical variables
label_encoders = {}
for col in ["Contains Link", "Contains Money Words", "Length", "Class"]:
    le = LabelEncoder()
    hamspam_df[col] = le.fit_transform(hamspam_df[col])
    label_encoders[col] = le

# Define features and target
X = hamspam_df[["Contains Link", "Contains Money Words", "Length"]]
y = hamspam_df["Class"]  # Encoded as 0 (Ham) and 1 (Spam)

# Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train Decision Tree model with limited depth
clf = DecisionTreeClassifier(max_depth=3, random_state=42)
clf.fit(X_train, y_train)

# Get probability scores
y_scores = clf.predict_proba(X_test)[:, 1]

# Compute ROC curve
fpr, tpr, _ = roc_curve(y_test, y_scores)
roc_auc = auc(fpr, tpr)

# Plot ROC curve
plt.figure(figsize=(6,6))
plt.plot(fpr, tpr, color="blue", lw=2, label=f"ROC curve (AUC = {roc_auc:.2f})")
plt.plot([0, 1], [0, 1], linestyle="--", color="gray", label="Random Guess")
plt.xlabel("False Positive Rate (FPR)")
plt.ylabel("True Positive Rate (TPR)")
plt.title("ROC Curve for Decision Tree Classifier")
plt.legend()
plt.grid(True)
plt.show()

# Output AUC score
roc_auc
```
